```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_dataset.csv")
print(df.head())
```

```python
#Mandatory steps
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#drop Unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","5-10")
```

```python
df.head()
```

GENDER DISTRIBUTION

```python
plt.figure(figsize=(4,6))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("GENDER DISTRIBUTION")
plt.show()
```

#from the above chart we have analyzied that:
#the number of females in the data is more than the number of males.

```python
#Groupby method
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
# Create the heatmap with adjusted parameters
sns.heatmap(gb,annot=True)
# Rotate x and y labels to prevent overlap
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
# Adjust layout to ensure everything fits
plt.tight_layout()
plt.title("Relationship B/W Parent Education vs kids score")
plt.show()
```

```python
#from the above chart we have concluded that the 
#education of the parents have the good impact on their scores
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship b/w Parents marital status vs kids score")
plt.show()
```

```python
#from the above chart we have analyzed that there is 
#no impact or we can say there is negligible impact on their studies
```

```python
groupA=df.loc[(df["EthnicGroup"]=='group A')].count()
groupB=df.loc[(df["EthnicGroup"]=='group B')].count()
groupC=df.loc[(df["EthnicGroup"]=='group C')].count()
groupD=df.loc[(df["EthnicGroup"]=='group D')].count()
groupE=df.loc[(df["EthnicGroup"]=='group E')].count()

label=["group A","group B","group C","group D","group E"]
myList=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(myList)
plt.pie(myList,labels=label,autopct="%1.2f")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
